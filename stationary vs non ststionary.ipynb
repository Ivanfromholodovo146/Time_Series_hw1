```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
import numpy as np
import matplotlib.pyplot as plt
```

```python
daily_total = pd.read_csv('/content/daily-total-female-births-in-cal.csv')
```

```python
daily_total.head()
```

```python
daily_total['Date'] =pd.to_datetime(daily_total['Date'])
```

```python
daily_total.info()
```

```python
daily_total.head()
```

```python
daily_total.set_index('Date', inplace = True)
```

```python
daily_total.plot()
```

```python
result = adfuller(daily_total['Count'])
test_statistic, pvalue,_ ,_ ,_ ,_ = result
if pvalue <= 0.05:
    print("The data is likely stationary")
else:
    print("The data is likely non-stationary")
```

```python
international = pd.read_csv('/content/international-airline-passengers.csv')
international.head()
```
